```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('times.csv')

# Crear una cuadrÃ­cula de 2x2 para las grÃ¡ficas
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# GrÃ¡fica 1: Boxplot para el tiempo de resoluciÃ³n por tipo de cubo
df.boxplot('Tiempo', by='TIpo', ax=axs[0, 0])
axs[0, 0].set_title('Boxplot del Tiempo de ResoluciÃ³n por Tipo de Cubo')

# GrÃ¡fica 2: Scatter plot para la cantidad de movimientos vs tiempo
axs[0, 1].scatter(df['Movimientos'], df['Tiempo'], c=pd.factorize(df['TIpo'])[0], cmap='viridis', s=50)
axs[0, 1].set_title('Cantidad de Movimientos vs Tiempo')
axs[0, 1].set_xlabel('Cantidad de Movimientos')
axs[0, 1].set_ylabel('Tiempo de ResoluciÃ³n')

# GrÃ¡fica 3: Histograma del tiempo de resoluciÃ³n
df['Tiempo'].hist(by=df['TIpo'], ax=axs[1, 0], bins=15)
axs[1, 0].set_title('Histograma del Tiempo de ResoluciÃ³n por Tipo de Cubo')

# GrÃ¡fica 4: Bar plot para el promedio de movimientos por tipo de jugador
df.groupby('Jugador')['Movimientos'].mean().plot(kind='bar', ax=axs[1, 1], color=['skyblue', 'lightcoral'])
axs[1, 1].set_title('Promedio de Movimientos por Tipo de Jugador')
axs[1, 1].set_ylabel('Promedio de Movimientos')

# Ajustar el diseÃ±o de las grÃ¡ficas
plt.tight_layout()

# Mostrar las grÃ¡ficas
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('times.csv')

# Graficar el tiempo de resoluciÃ³n segÃºn el tipo de jugador
plt.figure(figsize=(10, 6))

# Filtrar por tipo de jugador
jugadores = df['Jugador'].unique()

for jugador in jugadores:
    datos_jugador = df[df['Jugador'] == jugador]
    plt.scatter(datos_jugador['Movimientos'], datos_jugador['Tiempo'], label=jugador)

plt.title('Tiempo de resoluciÃ³n segÃºn la cantidad de movimientos')
plt.xlabel('Cantidad de movimientos')
plt.ylabel('Tiempo de resoluciÃ³n')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('times.csv')

# Graficar la cantidad de movimientos segÃºn el tipo de cubo
plt.figure(figsize=(10, 6))

# Filtrar por tipo de cubo
tipos_de_cubo = df['TIpo'].unique()

for tipo in tipos_de_cubo:
    datos_tipo_cubo = df[df['TIpo'] == tipo]
    plt.hist(datos_tipo_cubo['Movimientos'], bins=10, alpha=0.5, label=tipo)

plt.title('Cantidad de movimientos segÃºn el tipo de cubo')
plt.xlabel('Cantidad de movimientos')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('times.csv')

# Filtrar datos para los tipos de cubo 3x3 y 3x3 v2
df_3x3 = df[df['TIpo'].isin(['3x3', '3x3V2'])]

# Graficar la cantidad de movimientos para los tipos de cubo 3x3 y 3x3 v2
plt.figure(figsize=(10, 6))

for tipo in df_3x3['TIpo'].unique():
    datos_tipo_cubo = df_3x3[df_3x3['TIpo'] == tipo]
    plt.hist(datos_tipo_cubo['Movimientos'], bins=10, alpha=0.5, label=tipo)

plt.title('Cantidad de movimientos para los tipos de cubo 3x3 y 3x3 v2')
plt.xlabel('Cantidad de movimientos')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
